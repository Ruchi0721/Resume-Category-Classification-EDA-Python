##Cleaning the data
*   Hastags
*   Urls
*   Emojis  
*   Special characters   
*   punctuations





```python
df["Resume"][65]
```

```python
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText)
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
```

```python
df["Resume"] = df["Resume"].apply(lambda x : cleanResume(x))
# df["Resume"][3]
```

## Convert words into categorical values



```python
df.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df["Category"])
df["Category"] = le.transform(df["Category"])
df["Category"].nunique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = "english")
tfidf.fit(df["Resume"])
require_txt = tfidf.transform(df["Resume"])
# df
```

## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_text, y_train, y_test = train_test_split(require_txt,df["Category"],test_size = 0.2, random_state = 42)
```

## Train the model


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier

from sklearn.metrics import accuracy_score

#convert X_train and X_test into array

x_train = x_train.toarray() if hasattr(x_train, "toarray") else x_train
x_text = x_text.toarray() if hasattr(x_text,"toarray") else x_text
```

```python
# 1. Training KNN classifier
knn_model = OneVsRestClassifier(KNeighborsClassifier())
knn_model.fit(x_train,y_train)
y_pred_knn = knn_model.predict(x_text)
knn_accuracy = accuracy_score(y_test,y_pred_knn)
knn_accuracy
```

```python
     # 2. Training SVC classifier
svc_model = OneVsRestClassifier(SVC())
svc_model.fit(x_train,y_train)
y_pred_svc = svc_model.predict(x_text)
svc_accuracy = accuracy_score(y_test,y_pred_svc)
svc_accuracy
```

## Prediction on unseen data

```python
def pred(input_resume):

  # preprocess the input text
  clean_txt = cleanResume(input_resume)
  # Vectorize the cleaned text
  vectorized_text = tfidf.transform([clean_txt])
  # convert metrix into array
  vectorized_text = vectorized_text.toarray()

  # prdiction
  predicted_category = svc_model.predict(vectorized_text)

  # get the name of predicted class
  predicted_category_name = le.inverse_transform(predicted_category)

  return predicted_category_name[0]
```
